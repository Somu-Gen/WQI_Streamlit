```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

import plotly.express as px
```

```python
df=pd.read_csv("/content/Dec_2018.csv")
```

```python
df
```

```python
#Finding Mean
x_T = df['TDS'].mean()
x_TA = df['TA'].mean()
x_Ca = df['Ca'].mean()
x_Mg = df['Mg'].mean()
x_Na = df['Na'].mean()
x_K = df['K'].mean()
x_Cl = df['Cl'].mean()
x_Sulphate = df['Sulphate'].mean()
x_F = df['F'].mean()
x_pH = df['pH'].mean()
```

```python
#Filling NaN values with mean X
df['TDS'].fillna(x_T, inplace=True)
df['TA'].fillna(x_TA, inplace=True)
df['K'].fillna(x_K, inplace=True)
df['Ca'].fillna(x_Ca, inplace=True)
df['Mg'].fillna(x_Mg, inplace=True)
df['Na'].fillna(x_Na, inplace=True)
df['Cl'].fillna(x_Cl, inplace=True)
df['Sulphate'].fillna(x_Sulphate, inplace=True)
df['F'].fillna(x_F, inplace=True)
df['pH'].fillna(x_pH, inplace=True)
```

```python
df
```

```python
df.rename(columns={'Station name': 'Station_name','District / RO':'District/RO','Monitoring date':'Monitoring_date'}, inplace=True)
```

```python
df.drop(['S.No.','Station_name','District/RO','Monitoring_date'], axis=1)
```

```python
def wqi(vn,vi,s):
  qn=((vn-vi)/(s-vi))*100
  return qn


q_pH=wqi(df.pH,7,8.5)
df['q_pH']=q_pH

q_TA=wqi(df.TA,0,200)
df['q_TA']=q_TA

q_Cl=wqi(df.Cl,0,250)
df['q_Cl']=q_Cl


q_Ca=wqi(df.Ca,0,75)
df['q_Ca']=q_Ca

q_Mg=wqi(df.Mg,0,50)
df['q_Mg']=q_Mg


q_Sulphate=wqi(df.Sulphate ,0,250)
df['q_Sulphate']=q_Sulphate


q_Na=wqi(df.Na,0,200)
df['q_Na']=q_Na

q_TDS=wqi(df.TDS,0,500)
df['q_TDS']=q_TDS

q_K=wqi(df.K,0,12)
df['q_K']=q_K


q_F=wqi(df.F,0,1)
df['q_F']=q_F
```
